11. In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?

ANSWER

range(10) -range has 1 argument. The value 10 will represent the stop value, excluding that value.

range(0,10)- range has 2 arguments. The first value is start and second value is stop value(excluding that value).

range(0,10,1)- range has 3 arguments. the first value is start, second value is stop(excluding that value) and third value is the step between the values.
